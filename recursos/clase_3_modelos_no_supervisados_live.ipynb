Chuleta (cheatsheet) Algoritmos:

| Algoritmo                      | Tipo                               | Hiperparámetros principales                | ¿Requiere preprocesado?                          | ¿Cuándo usarlo?                                                            | Ventajas                                                         | Inconvenientes                                                                            |
| ------------------------------ | ---------------------------------- | ------------------------------------------ | ------------------------------------------------ | -------------------------------------------------------------------------- | ---------------------------------------------------------------- | ----------------------------------------------------------------------------------------- |
| **K-Means**                    | Particional                        | - n\_clusters (k)<br>- init<br>- max\_iter | Escalado de variables recomendado                | Cuando esperas grupos de forma "redondeada" y tienes idea del nº de grupos | Fácil de entender y rápido                                       | Sensible a outliers y al escalado; no detecta bien formas no esféricas; hay que indicar `k` |
| **DBSCAN**                     | Basado en densidad                 | - eps (ε)<br>- min\_samples (d)            | Escalado recomendable<br>(depende de la métrica) | Cuando no sabes el nº de grupos o hay ruido; grupos de cualquier forma     | Detecta ruido/outliers, no requiere `k`, grupos de cualquier forma | Puede fallar con densidades variables; definir `eps` es complicado                          |
| **HDBSCAN**                    | Basado en densidad jerárquica        | - min\_cluster\_size<br>- min\_samples     | Escalado recomendable   | Clusters de distinta densidad, encontrar ruido, sin saber `k`   | No requiere `k`, detecta outliers, mejor con densidades distintas que DBSCAN | Lento en grandes datasets, menos interpretabilidad, más complejo de configurar   |
| **Agglomerative (Jerárquico)** | Jerárquico                         | - linkage<br>- n\_clusters                 | Escalado recomendable                            | Cuando quieres visualizar la estructura jerárquica o no sabes k            | No requiere definir `k` a priori, visualización con dendrograma    | Costoso en memoria con muchos datos, sensible a la métrica y outliers                     |
| **Mean Shift**                 | Basado en densidad                 | - bandwidth                                | Escalado recomendable                            | Cuando no sabes el nº de clusters y buscas modos densos en los datos       | Detecta clusters de cualquier forma y cantidad                   | Lento en grandes datasets; elección de `bandwidth` difícil                                  |
| **Gaussian Mixture (GMM)**     | Modelos probabilísticos            | - n\_components<br>- covariance\_type      | Escalado muy recomendable                        | Cuando los grupos pueden solaparse o tienen forma elíptica                 | Permite solapamiento y probabilidades suaves de pertenencia      | Más lento, sensible a inicialización y outliers; hay que indicar nº de componentes        |
| **Spectral Clustering**        | Basado en grafos                   | - n\_clusters<br>- affinity                | Escalado recomendado                             | Cuando la forma de los clusters es compleja/no lineal                      | Puede encontrar clusters de formas complicadas                   | No escala bien a grandes datasets; requiere saber `k`; sensible a la afinidad elegida       |


### Reducción de la dimensionalidad

| Algoritmo                                     | Tipo                      | Hiperparámetros principales                                   | ¿Requiere preprocesado?                                | ¿Cuándo usarlo?                                                                               | Ventajas                                                                 | Inconvenientes                                                                          |
| --------------------------------------------- | ------------------------- | ------------------------------------------------------------- | ------------------------------------------------------ | --------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------ | --------------------------------------------------------------------------------------- |
| **PCA (Análisis de Componentes Principales)** | Lineal                    | - n\_components                                               | Escalado **muy recomendable**<br>(centrado en media 0) | Cuando los datos tienen correlación lineal, para visualización o preprocesado                 | Rápido, fácil de interpretar, útil para compresión y visualización 2D/3D | Solo capta relaciones lineales; puede perder interpretabilidad                          |
| **t-SNE**                                     | No lineal (proximidad)    | - n\_components<br>- perplexity<br>- learning\_rate           | Escalado recomendable                                  | Para visualizar datos complejos en 2D o 3D, clusters no lineales (imágenes, textos, genética) | Descubre estructuras complejas; visualización clara de grupos            | Muy lento para muchos datos (>10.000); resultados pueden variar; difícil de interpretar |
| **UMAP**                                      | No lineal (proximidad)    | - n\_neighbors<br>- min\_dist<br>- n\_components              | Escalado recomendable                                  | Alternativa rápida a t-SNE, para visualización o reducción previa al clustering               | Muy rápido, preserva bien la estructura de los datos                     | Los resultados pueden variar según los hiperparámetros; menos interpretable             |
| **Autoencoder**                               | No lineal (deep learning) | - arquitectura (capas, neuronas)<br>- epochs<br>- latent\_dim | Normalización o estandarización                        | Cuando hay muchos datos, relaciones muy complejas y se usan redes neuronales                  | Puede captar relaciones muy complejas; flexible                          | Requiere más datos, entrenamiento y conocimientos de deep learning                      |

````python
def comparar_clustering_escalado(
    df: pd.DataFrame, 
    columna_x: str, 
    columna_y: str, 
    columna_clase: str, 
    n_clusters: int = 3, 
    random_state: int = 42,
    mostrar_escalado: bool = True
) -> go.Figure:
    """
    Crea un subplot para comparar clustering K-means con diferentes métodos de escalado.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos
        columna_x (str): Nombre de la columna para el eje X
        columna_y (str): Nombre de la columna para el eje Y
        columna_clase (str): Nombre de la columna con las clases reales
        n_clusters (int): Número de clusters para K-means
        random_state (int): Semilla aleatoria para reproducibilidad
        mostrar_escalado (bool): Muestra los ejes aplicando el escalado
        
    Returns:
        go.Figure: Figura con los 4 subplots
    """
    
    # Crear figura con subplots 2x2
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=(
            "Clases Reales",
            "Clusters K-means (Sin escalar)",
            "Clusters K-means (StandardScaler)",
            "Clusters K-means (MinMaxScaler)"
        )
    )
    
    # Preparar datos para clustering (solo las dos columnas seleccionadas)
    X = df[[columna_x, columna_y]].values
    
    # Crear paleta de colores
    colores = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
    
    # 1. Gráfico de clases reales (arriba a la izquierda)
    clases_unicas = df[columna_clase].unique()
    for i, clase in enumerate(clases_unicas):
        mask = df[columna_clase] == clase
        fig.add_trace(
            go.Scatter(
                x=df.loc[mask, columna_x],
                y=df.loc[mask, columna_y],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Clase: {clase}',
                legendgroup='clases',
                showlegend=True
            ),
            row=1, col=1
        )
    
    # 2. K-means sin escalar (arriba a la derecha)
    kmeans_no_scaling = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters_no_scaling = kmeans_no_scaling.fit_predict(X)
    centroides_no_scaling = kmeans_no_scaling.cluster_centers_
    
    for i in range(n_clusters):
        mask = clusters_no_scaling == i
        fig.add_trace(
            go.Scatter(
                x=X[mask, 0],
                y=X[mask, 1],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Cluster {i} (Sin escalar)',
                legendgroup='no_scaling',
                showlegend=True
            ),
            row=1, col=2
        )
    
    # Añadir centroides sin escalar
    fig.add_trace(
        go.Scatter(
            x=centroides_no_scaling[:, 0],
            y=centroides_no_scaling[:, 1],
            mode='markers',
            marker=dict(
                color='black',
                size=12,
                symbol='x',
                line=dict(width=2)
            ),
            name='Centroides (Sin escalar)',
            legendgroup='no_scaling',
            showlegend=True
        ),
        row=1, col=2
    )
    
    # 3. K-means con StandardScaler (abajo a la izquierda)
    scaler_std = StandardScaler()
    X_std_scaled = scaler_std.fit_transform(X)
    kmeans_std = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters_std = kmeans_std.fit_predict(X_std_scaled)

    # Se recoge la variable que se mostrará en función de si está activado el parámetro de mostrar los valores escalados
    if mostrar_escalado:
        X_std_show = X_std_scaled
        centroides_std = kmeans_std.cluster_centers_
    else:
        X_std_show = X
        # Transformar los centroides de vuelta al espacio original
        centroides_std = scaler_std.inverse_transform(kmeans_std.cluster_centers_)

    for i in range(n_clusters):
        mask = clusters_std == i
        fig.add_trace(
            go.Scatter(
                x=X_std_show[mask, 0],  # Usar coordenadas originales para comparación
                y=X_std_show[mask, 1],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Cluster {i} (StandardScaler)',
                legendgroup='std_scaling',
                showlegend=True
            ),
            row=2, col=1
        )
    
    # Añadir centroides de StandardScaler
    fig.add_trace(
        go.Scatter(
            x=centroides_std[:, 0],
            y=centroides_std[:, 1],
            mode='markers',
            marker=dict(
                color='black',
                size=12,
                symbol='x',
                line=dict(width=2)
            ),
            name='Centroides (StandardScaler)',
            legendgroup='std_scaling',
            showlegend=True
        ),
        row=2, col=1
    )
    
    # 4. K-means con MinMaxScaler (abajo a la derecha)
    scaler_minmax = MinMaxScaler()
    X_minmax_scaled = scaler_minmax.fit_transform(X)
    kmeans_minmax = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters_minmax = kmeans_minmax.fit_predict(X_minmax_scaled)
    
    # Se recoge la variable que se mostrará en función de si está activado el parámetro de mostrar los valores escalados
    if mostrar_escalado:
        X_minmax_show = X_minmax_scaled
        centroides_minmax = kmeans_minmax.cluster_centers_
    else:
        X_minmax_show = X
        # Transformar los centroides de vuelta al espacio original
        centroides_minmax = scaler_minmax.inverse_transform(kmeans_minmax.cluster_centers_)
    
    for i in range(n_clusters):
        mask = clusters_minmax == i
        fig.add_trace(
            go.Scatter(
                x=X_minmax_show[mask, 0],  # Usar coordenadas originales para comparación
                y=X_minmax_show[mask, 1],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Cluster {i} (MinMaxScaler)',
                legendgroup='minmax_scaling',
                showlegend=True
            ),
            row=2, col=2
        )
    
    # Añadir centroides de MinMaxScaler
    fig.add_trace(
        go.Scatter(
            x=centroides_minmax[:, 0],
            y=centroides_minmax[:, 1],
            mode='markers',
            marker=dict(
                color='black',
                size=12,
                symbol='x',
                line=dict(width=2)
            ),
            name='Centroides (MinMaxScaler)',
            legendgroup='minmax_scaling',
            showlegend=True
        ),
        row=2, col=2
    )
    
    # Actualizar diseño y etiquetas
    fig.update_layout(
        title=f'Comparación de Métodos de Escalado para Clustering K-means ({columna_x} vs {columna_y})',
        height=800,
        width=1000,
        legend=dict(
            groupclick="toggleitem",
            tracegroupgap=5
        )
    )
    
    # Actualizar ejes con mismo título en todos los subplots
    fig.update_xaxes(title_text=columna_x, row=1, col=1)
    fig.update_xaxes(title_text=columna_x, row=1, col=2)
    fig.update_xaxes(title_text=columna_x, row=2, col=1)
    fig.update_xaxes(title_text=columna_x, row=2, col=2)
    
    fig.update_yaxes(title_text=columna_y, row=1, col=1)
    fig.update_yaxes(title_text=columna_y, row=1, col=2)
    fig.update_yaxes(title_text=columna_y, row=2, col=1)
    fig.update_yaxes(title_text=columna_y, row=2, col=2)
    
    return fig
````

````python
def plot_dendrogram(model, **kwargs):
    """
    Visualiza el dendrograma de un modelo de clustering jerárquico (AgglomerativeClustering).
    
    Esta función crea una matriz de enlace (linkage matrix) a partir de un modelo
    de clustering jerárquico ya entrenado y visualiza su dendrograma, mostrando
    cómo se fueron agrupando las observaciones.

    Fuente: https://scikit-learn.org/stable/auto_examples/cluster/plot_agglomerative_dendrogram.html
    
    Args:
        model (AgglomerativeClustering): Modelo de clustering jerárquico ya entrenado
            con compute_distances=True o con el método .fit() aplicado a los datos.
        **kwargs: Argumentos adicionales para pasar a la función dendrogram de scipy.
            Ejemplos: truncate_mode, p, leaf_font_size, color_threshold, etc.
    
    Returns:
        None: La función muestra el dendrograma pero no retorna ningún valor.
        
    Nota:
        El modelo debe tener los atributos children_ y distances_ disponibles
        (asegúrate de haber usado compute_distances=True al crear el modelo).
    """
    # Inicializamos un array para contar muestras bajo cada nodo
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    
    # Calculamos el número de muestras en cada fusión de clusters
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # Es un nodo hoja (muestra original)
            else:
                current_count += counts[child_idx - n_samples]  # Es un subcluster
        counts[i] = current_count
    
    # Creamos la matriz de enlace con la estructura requerida por scipy
    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)
    
    # Visualizamos el dendrograma con la matriz de enlace creada
    dendrogram(linkage_matrix, **kwargs)
````

````python
def visualizar_clusters_plotly(
    df: pd.DataFrame, 
    columna_x: str, 
    columna_y: str, 
    columna_cluster: str, 
    etiqueta_x: str = None, 
    etiqueta_y: str = None,
    titulo: str = 'Clusters resultantes',
    mostrar: bool = True
) -> go.Figure:
    """
    Visualiza clusters usando Plotly con un gráfico de dispersión interactivo.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos
        columna_x (str): Nombre de la columna para el eje X
        columna_y (str): Nombre de la columna para el eje Y
        columna_cluster (str): Nombre de la columna que contiene los clusters
        etiqueta_x (str, optional): Etiqueta personalizada para el eje X
        etiqueta_y (str, optional): Etiqueta personalizada para el eje Y
        titulo (str, optional): Título del gráfico
        mostrar (bool, optional): Si es True, muestra el gráfico
        
    Returns:
        go.Figure: Figura con el gráfico de dispersión
    """
    # Crear una copia para no modificar el dataframe original
    df_plot = df.copy()
    
    # Asegurar que la columna de cluster es string para mejor visualización
    df_plot[columna_cluster] = df_plot[columna_cluster].astype(str)
    
    # Usar las etiquetas proporcionadas o los nombres de columnas por defecto
    labels = {
        'x': etiqueta_x if etiqueta_x else columna_x,
        'y': etiqueta_y if etiqueta_y else columna_y
    }
    
    # Crear el gráfico con Plotly
    fig = px.scatter(
        df_plot, 
        x=columna_x, 
        y=columna_y, 
        color=columna_cluster,
        color_discrete_sequence=px.colors.qualitative.Set1,
        opacity=0.7,
        labels=labels,
        title=titulo
    )
    
    # Actualizar diseño del gráfico
    fig.update_layout(
        width=800, 
        height=500,
        coloraxis_colorbar=dict(
            title='Cluster'
        )
    )
        
    return fig
````

````python
def visualizar_distribucion_clusters(
    df: pd.DataFrame, 
    columna_cluster: str, 
    columna_clase: str, 
    list_nombres_clases: list = None, 
    colores: list = None, 
    titulo: str = 'Distribución de clases reales en cada cluster',
) -> go.Figure:
    """
    Crea una visualización de la distribución de clases en cada cluster con Plotly.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos
        columna_cluster (str): Nombre de la columna que contiene los clusters
        columna_clase (str): Nombre de la columna que contiene las clases reales
        list_nombres_clases (list, optional): Lista con los nombres de las clases
        colores (list, optional): Lista de colores en formato hexadecimal
        titulo (str, optional): Título del gráfico
        
    Returns:
        go.Figure: Figura con los gráficos de distribución
    """
    
    # Si no se proporcionan nombres de clases, usamos los valores únicos de la columna clase
    if list_nombres_clases is None:
        list_nombres_clases = df[columna_clase].unique()
        
    # Si no se proporcionan colores, usamos la paleta Set1
    if colores is None:
        colores = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
    
    # Obtener la distribución de clases en cada cluster (valores absolutos)
    cluster_class_counts_abs = pd.crosstab(
        df[columna_cluster], 
        df[columna_clase]
    )
    
    # Calcular la distribución porcentual de clases en cada cluster
    cluster_class_counts = cluster_class_counts_abs.div(cluster_class_counts_abs.sum(axis=1), axis=0) * 100

    # Crear figura con subplots
    fig = make_subplots(
        rows=1, cols=2, 
        subplot_titles=('Distribución absoluta', 'Distribución porcentual (%)'),
        specs=[[{'type': 'bar'}, {'type': 'bar'}]]
    )

    # Añadir barras para cada clase en valores absolutos
    for i, clase in enumerate(list_nombres_clases):
        if clase in cluster_class_counts_abs.columns:
            fig.add_trace(
                go.Bar(
                    x=[f'Cluster {cluster}' for cluster in cluster_class_counts_abs.index],
                    y=cluster_class_counts_abs[clase],
                    name=clase,
                    marker_color=colores[i % len(colores)],
                    text=cluster_class_counts_abs[clase],
                    textposition='auto',
                ),
                row=1, col=1
            )

    # Añadir barras para cada clase en porcentajes
    for i, clase in enumerate(list_nombres_clases):
        if clase in cluster_class_counts.columns:
            fig.add_trace(
                go.Bar(
                    x=[f'Cluster {cluster}' for cluster in cluster_class_counts.index],
                    y=cluster_class_counts[clase],
                    name=clase,
                    marker_color=colores[i % len(colores)],
                    text=[f'{val:.1f}%' for val in cluster_class_counts[clase]],
                    textposition='auto',
                    showlegend=False
                ),
                row=1, col=2
            )

    # Actualizar layout
    fig.update_layout(
        title=titulo,
        barmode='group',
        height=500,
        width=1000,
        legend=dict(
            x=0.5,
            y=1.15,
            xanchor='center',
            orientation='h'
        )
    )

    fig.update_yaxes(title_text='Número de observaciones', row=1, col=1)
    fig.update_yaxes(title_text='Porcentaje (%)', row=1, col=2)
        
    return fig
````

````python
def visualizar_pca_varianza(
    datos: pd.DataFrame, 
    aplicar_escalado: bool = True,
    umbral_varianza: float = 0.95,
    random_state: int = 42,
    titulo: str = "Scree plot y varianza acumulada (PCA)"
) -> go.Figure:
    """
    Visualiza la varianza explicada por cada componente principal y la varianza acumulada.
    
    Args:
        datos (pd.DataFrame): DataFrame con los datos a los que aplicar PCA
        aplicar_escalado (bool): Si True, aplica StandardScaler antes de PCA
        umbral_varianza (float): Umbral para marcar en el gráfico (ej: 0.95 para 95%)
        random_state (int): Semilla para reproducibilidad
        titulo (str): Título del gráfico
        
    Returns:
        go.Figure: Figura de Plotly con el gráfico de varianza de PCA
    """
    # Preparamos los datos
    if aplicar_escalado:
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(datos)
    else:
        X_scaled = datos.values
    
    # Aplicamos PCA
    pca = PCA(random_state=random_state)
    pca.fit(X_scaled)
    
    # Obtenemos los datos de varianza
    varianza = pca.explained_variance_ratio_
    varianza_acum = np.cumsum(varianza)
    componentes = list(range(1, len(varianza)+1))
    
    # Creamos la figura con dos ejes Y
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    # Añadimos el gráfico de barras para la varianza individual
    fig.add_trace(
        go.Bar(
            x=componentes,
            y=varianza,
            name="Varianza por componente",
            marker_color='rgb(55, 83, 109)',
            opacity=0.7,
            hovertemplate='Componente %{x}<br>Varianza explicada: %{y:.3f}<extra></extra>'
        ),
        secondary_y=False
    )
    
    # Añadimos la línea para la varianza acumulada
    fig.add_trace(
        go.Scatter(
            x=componentes,
            y=varianza_acum,
            name="Varianza acumulada",
            mode='lines+markers',
            marker=dict(color='red'),
            line=dict(width=3),
            hovertemplate='Componente %{x}<br>Varianza acumulada: %{y:.3f}<extra></extra>'
        ),
        secondary_y=True
    )
    
    # Configuramos los ejes y el diseño
    fig.update_layout(
        title=titulo,
        xaxis=dict(
            title="Número de componentes principales",
            tickmode='linear',
            tick0=1,
            dtick=1
        ),
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="center",
            x=0.5
        ),
        hovermode="x unified",
        width=900,
        height=600,
        template="plotly_white",
    )
    
    # Configuramos los títulos de los ejes Y
    fig.update_yaxes(title_text="Proporción de varianza explicada", secondary_y=False)
    fig.update_yaxes(
        title_text="Varianza acumulada", 
        secondary_y=True,
        range=[0, 1.05],
        tickformat=".0%"
    )
    
    # Añadimos línea de referencia en el umbral de varianza acumulada
    component_threshold = next((i for i, v in enumerate(varianza_acum) if v >= umbral_varianza), 
                              len(varianza_acum))
    
    if component_threshold < len(varianza_acum):
        fig.add_vline(
            x=component_threshold + 1,  # +1 porque los componentes empiezan en 1
            line_dash="dash",
            line_color="green",
            annotation_text=f"{umbral_varianza*100:.0f}% de varianza con {component_threshold + 1} componentes",
            annotation_position="top right"
        )
    
    return fig
````

````python
def visualizar_proyeccion_2d(
    df: pd.DataFrame, 
    columna_x: str = 'Componente 1', 
    columna_y: str = 'Componente 2', 
    columna_color: str = None,
    etiqueta_x: str = None, 
    etiqueta_y: str = None,
    titulo: str = 'Proyección 2D',
    metodo: str = 'PCA',
    paleta_colores: list = None,
    ancho: int = 800,
    alto: int = 600,
    tamano_punto: int = 8,
    mostrar_leyenda: bool = True,
    opacidad: float = 0.8
) -> go.Figure:
    """
    Visualiza proyecciones 2D de técnicas de reducción de dimensionalidad como PCA, UMAP o t-SNE.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos proyectados (debe contener al menos las columnas columna_x y columna_y)
        columna_x (str): Nombre de la columna para el eje X (por defecto: 'Componente 1')
        columna_y (str): Nombre de la columna para el eje Y (por defecto: 'Componente 2')
        columna_color (str): Nombre de la columna para colorear los puntos (opcional)
        etiqueta_x (str): Etiqueta personalizada para el eje X
        etiqueta_y (str): Etiqueta personalizada para el eje Y
        titulo (str): Título del gráfico
        metodo (str): Método de reducción de dimensionalidad ('PCA', 'UMAP', 't-SNE', etc.)
        paleta_colores (list): Lista personalizada de colores (si None, usa la paleta Set1)
        ancho (int): Ancho del gráfico en píxeles
        alto (int): Altura del gráfico en píxeles
        tamano_punto (int): Tamaño de los puntos en el gráfico
        mostrar_leyenda (bool): Si es True, muestra la leyenda
        opacidad (float): Opacidad de los puntos (0-1)
        
    Returns:
        go.Figure: Figura de Plotly con la proyección 2D
    """
    # Verificar que las columnas existen en el dataframe
    for col in [columna_x, columna_y]:
        if col not in df.columns:
            raise ValueError(f"La columna '{col}' no existe en el DataFrame proporcionado")
    
    if columna_color and columna_color not in df.columns:
        raise ValueError(f"La columna de color '{columna_color}' no existe en el DataFrame")
    
    # Crear una copia para no modificar el dataframe original
    df_plot = df.copy()
    
    # Asegurar que la columna de cluster sea del tipo adecuado para plotly
    if columna_color and 'cluster' in columna_color.lower():
        # Convertir columna a tipo categórico para evitar problemas con valores numéricos
        df_plot[columna_color] = df_plot[columna_color].astype(str)
    
    # Determinar las etiquetas de los ejes
    if not etiqueta_x:
        if 'PC' in columna_x:
            etiqueta_x = f'Componente {columna_x[2:]}'
        else:
            etiqueta_x = columna_x
            
    if not etiqueta_y:
        if 'PC' in columna_y:
            etiqueta_y = f'Componente {columna_y[2:]}'
        else:
            etiqueta_y = columna_y
    
    # Configurar el título completo
    titulo_completo = f'{titulo} ({metodo})'
    
    # Configurar las opciones para el scatter plot
    scatter_kwargs = {
        'x': columna_x,
        'y': columna_y,
        'opacity': opacidad,
        'title': titulo_completo,
        'labels': {columna_x: etiqueta_x, columna_y: etiqueta_y}
    }
    
    # Si hay una columna para colorear los puntos
    if columna_color:
        scatter_kwargs['color'] = columna_color
        
        # Verificar si es numérica para usar color continuo o discreto
        if pd.api.types.is_numeric_dtype(df_plot[columna_color]) and 'cluster' not in columna_color.lower():
            # Si es numérica y no es un cluster, usar escala continua
            scatter_kwargs['color_continuous_scale'] = 'Viridis'
        else:
            # Si es categórica o un cluster, usar colores discretos
            # Usar la paleta personalizada o la predeterminada
            if paleta_colores:
                scatter_kwargs['color_discrete_sequence'] = paleta_colores
            else:
                scatter_kwargs['color_discrete_sequence'] = px.colors.qualitative.Set1
            
            # No usar 'identity' que puede causar problemas con valores numéricos
            scatter_kwargs.pop('color_discrete_map', None)
    else:
        # Si no hay una columna de color, usar un único color
        scatter_kwargs['color_discrete_sequence'] = ['#636EFA']
    
    # Crear el gráfico con Plotly Express
    fig = px.scatter(df_plot, **scatter_kwargs)
    
    # Actualizar el diseño del gráfico
    layout_updates = {
        'width': ancho,
        'height': alto,
        'template': 'plotly_white',
        'xaxis_title': etiqueta_x,
        'yaxis_title': etiqueta_y,
        'showlegend': mostrar_leyenda
    }
    
    # Si hay una columna de color, ajustar el título de la leyenda
    if columna_color:
        layout_updates['legend_title_text'] = columna_color.capitalize()
    
    # Actualizar el tamaño de los marcadores
    fig.update_traces(marker=dict(size=tamano_punto))
    
    # Aplicar las actualizaciones de diseño
    fig.update_layout(**layout_updates)
    
    return fig

def visualizar_proyeccion_3d(
    df: pd.DataFrame, 
    columna_x: str = 'Componente 1', 
    columna_y: str = 'Componente 2', 
    columna_z: str = 'Componente 3',
    columna_color: str = None,
    etiqueta_x: str = None, 
    etiqueta_y: str = None,
    etiqueta_z: str = None,
    titulo: str = 'Proyección 3D',
    metodo: str = 'PCA',
    paleta_colores: list = None,
    ancho: int = 900,
    alto: int = 700,
    tamano_punto: int = 6,
    mostrar_leyenda: bool = True,
    opacidad: float = 0.8,
    angulo_camara: dict = None
) -> go.Figure:
    """
    Visualiza proyecciones 3D de técnicas de reducción de dimensionalidad como PCA, UMAP o t-SNE.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos proyectados 
        columna_x (str): Nombre de la columna para el eje X (por defecto: 'Componente 1')
        columna_y (str): Nombre de la columna para el eje Y (por defecto: 'Componente 2')
        columna_z (str): Nombre de la columna para el eje Z (por defecto: 'Componente 3')
        columna_color (str): Nombre de la columna para colorear los puntos (opcional)
        etiqueta_x (str): Etiqueta personalizada para el eje X
        etiqueta_y (str): Etiqueta personalizada para el eje Y
        etiqueta_z (str): Etiqueta personalizada para el eje Z
        titulo (str): Título del gráfico
        metodo (str): Método de reducción de dimensionalidad ('PCA', 'UMAP', 't-SNE', etc.)
        paleta_colores (list): Lista personalizada de colores (si None, usa la paleta Set1)
        ancho (int): Ancho del gráfico en píxeles
        alto (int): Altura del gráfico en píxeles
        tamano_punto (int): Tamaño de los puntos en el gráfico
        mostrar_leyenda (bool): Si es True, muestra la leyenda
        opacidad (float): Opacidad de los puntos (0-1)
        angulo_camara (dict): Diccionario con los ángulos de la cámara (eye, center, up)
        
    Returns:
        go.Figure: Figura de Plotly con la proyección 3D
    """
    # Verificar que las columnas existen en el dataframe
    for col in [columna_x, columna_y, columna_z]:
        if col not in df.columns:
            raise ValueError(f"La columna '{col}' no existe en el DataFrame proporcionado")
    
    if columna_color and columna_color not in df.columns:
        raise ValueError(f"La columna de color '{columna_color}' no existe en el DataFrame")
    
    # Crear una copia para no modificar el dataframe original
    df_plot = df.copy()
    
    # Determinar las etiquetas de los ejes
    def generar_etiqueta(columna, etiqueta_predefinida):
        if etiqueta_predefinida:
            return etiqueta_predefinida
        elif 'PC' in columna:
            return f'Componente {columna[2:]}'
        else:
            return columna
    
    etiqueta_x = generar_etiqueta(columna_x, etiqueta_x)
    etiqueta_y = generar_etiqueta(columna_y, etiqueta_y)
    etiqueta_z = generar_etiqueta(columna_z, etiqueta_z)
    
    # Configurar el título completo
    titulo_completo = f'{titulo} ({metodo})'
    
    # Crear la figura 3D
    fig = go.Figure()
    
    # Determinar los colores para los puntos
    if columna_color:
        # Si es numérica, usamos escala continua
        if pd.api.types.is_numeric_dtype(df_plot[columna_color]):
            fig.add_trace(
                go.Scatter3d(
                    x=df_plot[columna_x],
                    y=df_plot[columna_y],
                    z=df_plot[columna_z],
                    mode='markers',
                    marker=dict(
                        size=tamano_punto,
                        color=df_plot[columna_color],
                        opacity=opacidad,
                        colorscale='Viridis',
                        showscale=True,
                        colorbar=dict(title=columna_color)
                    ),
                    hovertemplate=(
                        f"{etiqueta_x}: %{{x:.2f}}<br>"
                        f"{etiqueta_y}: %{{y:.2f}}<br>"
                        f"{etiqueta_z}: %{{z:.2f}}<br>"
                        f"{columna_color}: %{{marker.color}}<extra></extra>"
                    )
                )
            )
        else:
            # Para variables categóricas, usamos colores discretos
            categorias = df_plot[columna_color].unique()
            
            # Seleccionar paleta de colores
            if paleta_colores:
                colores = paleta_colores
            else:
                colores = px.colors.qualitative.Set1
            
            # Crear un trace para cada categoría
            for i, categoria in enumerate(categorias):
                mask = df_plot[columna_color] == categoria
                fig.add_trace(
                    go.Scatter3d(
                        x=df_plot.loc[mask, columna_x],
                        y=df_plot.loc[mask, columna_y],
                        z=df_plot.loc[mask, columna_z],
                        mode='markers',
                        marker=dict(
                            size=tamano_punto,
                            color=colores[i % len(colores)],
                            opacity=opacidad
                        ),
                        name=str(categoria),
                        hovertemplate=(
                            f"{etiqueta_x}: %{{x:.2f}}<br>"
                            f"{etiqueta_y}: %{{y:.2f}}<br>"
                            f"{etiqueta_z}: %{{z:.2f}}<br>"
                            f"{columna_color}: {categoria}<extra></extra>"
                        )
                    )
                )
    else:
        # Si no hay columna de color, usar un único color
        fig.add_trace(
            go.Scatter3d(
                x=df_plot[columna_x],
                y=df_plot[columna_y],
                z=df_plot[columna_z],
                mode='markers',
                marker=dict(
                    size=tamano_punto,
                    color='#636EFA',
                    opacity=opacidad
                ),
                hovertemplate=(
                    f"{etiqueta_x}: %{{x:.2f}}<br>"
                    f"{etiqueta_y}: %{{y:.2f}}<br>"
                    f"{etiqueta_z}: %{{z:.2f}}<extra></extra>"
                )
            )
        )
    
    # Actualizar el diseño del gráfico
    layout_updates = {
        'width': ancho,
        'height': alto,
        'title': titulo_completo,
        'scene': dict(
            xaxis_title=etiqueta_x,
            yaxis_title=etiqueta_y,
            zaxis_title=etiqueta_z,
            # Configurar aspecto para que sea cúbico
            aspectmode='cube'
        ),
        'template': 'plotly_white',
        'showlegend': mostrar_leyenda
    }
    
    # Configurar ángulo de cámara si se proporciona
    if angulo_camara:
        layout_updates['scene']['camera'] = angulo_camara
    
    # Aplicar las actualizaciones de diseño
    fig.update_layout(**layout_updates)
    
    return fig
````

### Creación del modelo
+ En este caso utilizaremos las 4 variables que se utilizan para medir los pétalos y sépalos de cada flor con el objetivo de encontrar los diferentes grupos que se puedan formar.
+ Dentro del paradigma de muchos modelos de clustering es responder a la pregunta "¿Cuántos clusters se podrán encontrar en los datos?", ya que hay que definir este número previamente.
+ K-Means no te dice cuál es el número de clusters más óptimo.
+ No utilizaremos el método de `train_test_split` porque normalmente los modelos de clustering no se suelen evaluar.

```python
from sklearn.cluster import KMeans
```

+ `n_clusters`: representa el valor `k`, que es el número de clusters.
+ `init`: método de inicialización de los centroides, el valor más común (y por defecto) es `k-means++`. También puede ser `random`.
+ `max_iter`: número de iteraciones máximas que hará hasta encontrar los centroides más óptimos.

### Evaluación del modelo
Al ser un modelo no supervisado, nunca sabremos si es la buena asignación de clusters o no, entonces hay técnicas y métricas que ayudan a ver si los puntos (cada una de las observaciones) de cada cluster se encuentran cerca de su centroide o si los propios clusters están agrupados entre sí y separados del resto.

#### Elbow Method
La manera tradicional de encontrar el número óptimo de clusters es el famoso Elbow Method (Método del Codo), a través de un gráfico viendo cómo decrede un valor llamado `inertia`.

$$\text{Inertia} = \sum_{i=1}^{n} \min_{\mu_j \in C} |x_i - \mu_j|^2$$

Donde:
+ $x_i$ representa cada punto de datos
+ $\mu_j$ representa cada centroide
+ $C$ es el conjunto de todos los centroides
+ $|x_i - \mu_j|^2$ es el cuadrado de la distancia euclidiana entre el punto $x_i$ y el centroide $\mu_j$
+ $\min_{\mu_j \in C}$ indica que para cada punto tomamos la distancia mínima a cualquiera de los centroides

````python
# Definimos un valor de inertia
inertia = []

# Generamos un rango de valores de k (número de clusters) para probar
k_range = range(1, 11)

# Iteramos sobre el rango de k y calculamos la inercia para cada valor
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_iris)

    inertia.append(km.inertia_)

# Visualizamos el método del codo (Elbow Method)
plt.figure(figsize=(7, 7))
plt.plot(k_range, inertia, marker='o')
plt.show()
````

#### Silhouette method
El Silhouette Score mide qué tan bien separado y compacto está cada grupo (cluster) después de aplicar un algoritmo de clustering como K-Means, este valor irá entre 0 y 1. También, es una manera matemática de elegir el mejor número de clusters en algoritmos como K-Means.

Cómo se mide:
+ Puntos cerca de su propio centroide y lejos de otros → Silhouette ≈ 1
+ Puntos entre clusters (frontera) → Silhouette ≈ 0
+ Puntos más cerca de otro cluster → Silhouette < 0

Importaremos la función `silhouett_score` para obtener este valor:
```python
from sklearn.metrics import silhouette_score
```

````python
# Utilizando el método de la silueta para determinar el número óptimo de clusters
scores = []
k_values = range(2, 11)

# Calculamos el silhouette score para cada valor de k
for k in k_values:
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(X_iris)

    score = silhouette_score(X_iris, labels)

    scores.append(score)
    print(f"Silhouette score para k={k}: {score}")

# Visualizamos el silhouette score para cada valor de k
plt.figure(figsize=(7,4))
plt.plot(k_values, scores, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Silhouette score')
plt.title('Silhouette Method para elegir k')
plt.show()
````

### Importancia del escalado o normalizar los datos

El escalado de datos es crucial en clustering por estos motivos:
+ **Evita el sesgo por magnitudes**: Algoritmos como K-means son sensibles a variables con diferentes escalas, dando más peso a las que tienen valores más grandes
+ **Mejora cálculos de distancia**: Las métricas de distancia funcionan correctamente cuando todas las variables tienen escalas comparables
+ **Acelera la convergencia**: Los algoritmos convergen más rápido y de forma más estable con datos escalados

Principales técnicas:
+ `StandardScaler` (estandarización):
  + Transforma variables para tener media 0 y desviación estándar 1
  + Ideal para: Distribuciones aproximadamente normales o datos con outliers
  + Fórmula: `z = (x - μ)/σ`
$$z = \frac{x - \mu}{\sigma}$$

+ `MinMaxScaler` (normalizar):
  + Escala datos al rango [0,1]
  + Ideal para: Preservar relaciones exactas entre valores o distribuciones no gaussianas
  + Fórmula: `x_scaled = (x - x_min)/(x_max - x_min)`

$$x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}$$
